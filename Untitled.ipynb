```python
import datetime
import math
import pandas as pd
```

```python
london_tweets = pd.read_csv("./data/london_tweets.csv", )
```

```python
london_tweets.head()
```

```python
borough_to_cluster = pd.read_csv("./data/borough_to_cluster.csv")
```

```python
borough_to_cluster.head()
```

```python
tweets_by_borough_cluster = london_tweets.merge(borough_to_cluster, left_on='LOCATION', right_on='LOCATION', how='outer')
tweets_by_borough_cluster.head()
```

```python
tweets_by_borough_cluster.groupby("Cluster")["id"].agg("count")
```

```python
tweets_by_borough_cluster[tweets_by_borough_cluster.LOCATION == "City of London"].shape
```

```python
tweets_grouped_by_borough = london_tweets.groupby("LOCATION")["id"].agg("count").reset_index()
```

```python
tweets_grouped_by_borough.columns = ['borough', 'tweets']
tweets_grouped_by_borough.head()
```

```python
tweets_grouped_by_borough.to_csv("./data/tweets_grouped_by_borough.csv", index=False)
```

```python
tweets_grouped_by_borough.sort_values("tweets")
```
